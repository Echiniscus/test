# Import des librairies

```python
import pandas as pd
import seaborn as sns
```

# Chargement des données

```python
df=pd.read_csv('trainingData.tsv',delimiter='\t',encoding='utf-8')
print(list(df.columns.values)) 
print(df.tail(35)) 
```

# Réalisation de graphiques pour la description de chaque variable

## Target en fonction de l'âge

```python
sns.set(style="whitegrid")
sns.catplot(x="target", y="age", kind="swarm",data=df)
```

## Répartition des douleurs thoraciques en fonction de la target

```python
count={'douleur autre':0,'asymptomatique':0,'douleur thoracique typique':0,'douleur thoracique atypique':0}
for i in df['douleurThoracique']:
    count[i]+=1

print(count)

sns.catplot(y="target", hue="douleurThoracique", kind="count",palette="pastel", edgecolor=".6",data=df)
```

## Répartion des sexes en fonction de la target

```python
count={'femme':0,'homme':0}
for i in df['sexe']:
    count[i]+=1

print(count)

sns.catplot(y="sexe", hue="target", kind="count",palette="pastel", edgecolor=".6",data=df)
```

## Target en fonction de la PA

```python
sns.set(style="whitegrid")
sns.catplot(x="target", y="PA", kind="swarm",data=df)
```

## Target en fonction du cholesterol

```python
sns.set(style="whitegrid")
sns.catplot(x="target", y="cholesterol", kind="swarm",data=df)
```

## Target en fonction de la FCmax

```python
sns.set(style="whitegrid")
sns.catplot(x="target", y="FCmax", kind="swarm",data=df)
```

## Target en fonction de STdepression

```python
sns.set(style="whitegrid")
sns.catplot(x="target", y="STdepression", kind="swarm",data=df)
```

## Target en fonction du nombre de vaisseaux visibles

```python
sns.set(style="whitegrid")
sns.catplot(x="target", y="VaisseauxVisibles", kind="swarm",data=df)
```

## Répartition des Targets en fonction de la Glycémie à jeun

```python
count={'normale':0,'anormale':0}
for i in df['GlycemieAjeun']:
    count[i]+=1

print(count)

sns.catplot(y="target", hue="GlycemieAjeun", kind="count",palette="pastel", edgecolor=".6",data=df)
```

## Répartition des Targets en fonction des ECG normaux ou anormaux

```python
count={'anormal':0,'normal':0}
for i in df['ECG']:
    count[i]+=1

print(count)

sns.catplot(y="ECG", hue="target", kind="count",palette="pastel", edgecolor=".6",data=df)
```

## Répartition des Targets en fonction des douleurs à l'effort

```python
count={0:0,1:0}
for i in df['DouleurEffort']:
    count[i]+=1
    
print(count)

sns.catplot(y="target", hue="DouleurEffort", kind="count",palette="pastel", edgecolor=".6",data=df);
```

## Répartition des Targets en fonction de la STpente

```python
count={0:0,1:0,2:0}
for i in df['STpente']:
    count[i]+=1

print(count)

sns.catplot(y="target", hue="STpente", kind="count",palette="pastel", edgecolor=".6",data=df);
```
